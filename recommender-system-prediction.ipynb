#### train a model and save it
supported models are: 'SVD', 'KNN', 'NMF'

```python
data = get_data(from_surprise=True)
train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)
# train NMF model
model_NMF, cost_time = get_trained_model(algo='NMF', 
                                     model_kwargs={'n_factors':15, 'n_epochs': 50, 'verbose':False}, 
                                     train_set=train_set, 
                                     save_path='NMF.model')
# train SVD model
model_SVD, cost_time = get_trained_model(algo='SVD', 
                                     model_kwargs={}, 
                                     train_set=train_set, 
                                     save_path='SVD.model')
# train KNN model
model_KNN, cost_time = get_trained_model(algo='KNN', 
                                     model_kwargs={'k': 40, 'min_k': 1, 'sim_options': {'user_based': False, 'name': 'pearson'}}, 
                                     train_set=train_set, 
                                     save_path='KNN.model')
```
